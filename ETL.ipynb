## ETL

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('DATASETS/calidad_aire_bsas')
df.head()
```

```python
#La hora ya está incluída en el formato fecha por lo que vamos a eliminar esa columna
df.drop(['HORA'], axis=1,inplace=True)
df.head()
```

```python
COLUMNAS = ['FECHA','CO_CENTENARIO','NO2_CENTENARIO','PM10_CENTENARIO']
dfcentenario= df[COLUMNAS]
dfcentenario.head()
```

```python
dfcentenario.insert(loc=2,column='Unidad',value='ppm')
dfcentenario.insert(loc=4,column='unid',value='ppb')
dfcentenario.insert(loc=6,column='unidad',value='µg/m3')
```

```python
dfcentenario.head()
```

```python
columnas = ['FECHA','CO_CENTENARIO','Unidad']
dfcentenario1 = dfcentenario[columnas]
dfcentenario1.rename(columns={'CO_CENTENARIO': 'Medida'},inplace=True)
dfcentenario1.insert(loc=3, column = 'Id_Contaminante', value = '1')
dfcentenario1.head()
```

```python
columnas = ['FECHA','NO2_CENTENARIO','unid']
dfcentenario2 = dfcentenario[columnas]
dfcentenario2.rename(columns={'NO2_CENTENARIO': 'Medida', 'unid':'Unidad'},inplace=True)
dfcentenario2.insert(loc=3, column = 'Id_Contaminante', value = '2')
dfcentenario2.head()
```

```python
columnas = ['FECHA','PM10_CENTENARIO','unidad']
dfcentenario3 = dfcentenario[columnas]
dfcentenario3.rename(columns={'PM10_CENTENARIO': 'Medida', 'unidad':'Unidad'},inplace=True)
dfcentenario3.insert(loc=3, column = 'Id_Contaminante', value = '3')
dfcentenario3.head()
```

### Apilo todas las mediciones realizadas por la estacion centenario en un solo dataframe

```python
dfCENTENARIO = pd.concat([dfcentenario1,dfcentenario2,dfcentenario3])
dfCENTENARIO.insert(loc=4,column='Id_Estación',value='1')
dfCENTENARIO.head()
```
