```python
# Como se puede ver en la descripción de los contaminantes, es probable que las estaciones de palermo no hayan detectado particulas por lo que no registran valores.
# Lo que voy a hacer es reemplazar los valores nulos por 0
df = df.fillna(0)
df.head()
```

```python
# También reemplazaremos los 's/d' (sin datos) por 0
df.replace({'s/d' : 0}, inplace=True)
df.head()
```

```python
#Antes de guardar el nuevo dataframe y cargarlo a la base de datos de MySQL, hago un repaso general del mismo
df.info() 
```

```python
# Me aseguro de asignar el type correcto a cada columna
columnas_a_cambiar = ['CO_CENTENARIO', 'NO2_CENTENARIO', 'PM10_CENTENARIO', 'CO_CORDOBA', 'NO2_CORDOBA','PM10_CORDOBA', 'CO_LA_BOCA','NO2_LA_BOCA','PM10_LA_BOCA','CO_PALERMO','NO2_PALERMO','PM10_PALERMO']
df[columnas_a_cambiar] = df[columnas_a_cambiar].astype(float)
df.head()
```

```python
# Hay valores menores a 0.05 y están escritos como '<0.05', por lo tanto al intentar convertirlos a float me marca error.
# Reemplazo ese valor por 0.05
df = df.replace('<', '', regex=True)
df.info() #ahora las columnas 'object' deberían ser 'int'
```

```python
df.replace({'#REF!': 0},inplace=True)
df.info()
```

```python
df.replace({'S/d':0}, inplace=True)
df.info()
```

```python
#Vuelvo a ejecutar el código para cambiar el tipo de dato de las columnas
columnas_a_cambiar = ['CO_CENTENARIO', 'NO2_CENTENARIO', 'PM10_CENTENARIO', 'CO_CORDOBA', 'NO2_CORDOBA','PM10_CORDOBA', 'CO_LA_BOCA','NO2_LA_BOCA','PM10_LA_BOCA','CO_PALERMO','NO2_PALERMO','PM10_PALERMO']
df[columnas_a_cambiar] = df[columnas_a_cambiar].astype(float)
df.info()
```

```python
#Exportamos el archivo como csv para cargarlo en la base de datos
df.to_csv('DATASETS/calidad_aire_bsas', index=False)
```
